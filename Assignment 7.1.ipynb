```python
# Importing required package
import pandas as pd
```

```python
# Loading the dataset
df = pd.read_csv(
    "Canada.xlsx - Canada by Citizenship.csv", 
    skiprows=20
)
```

```python
# Displaying the first few rows of the raw dataset
df.head()
```

```python
# Renaming the columns
df.rename(columns={
    "OdName": "Country",
    "AreaName": "Continent",
    "RegName": "Region"
}, inplace=True)
```

```python
# Displaying the first few rows after renaming
df.head()
```

```python
# Dropping unnecessary columns
cols_to_drop = [
    "Type", "Coverage", "AREA", "REG", "DEV", "Area", "Reg", "Dev"
]
df.drop(columns=cols_to_drop, inplace=True, errors="ignore")
```

```python
# Setting Country as index
df.set_index("Country", inplace=True)
```

```python
# Identifying year columns
years = list(map(str, range(1980, 2014)))
```

```python
# Ensuring year columns are numeric
df[years] = df[years].apply(pd.to_numeric, errors="coerce")
```

```python
# Calculating the total immigration per country
df["Total"] = df[years].sum(axis=1)
```

```python
# Reordering columns
final_columns = ["Continent", "Region", "DevName"] + years + ["Total"]
df = df[final_columns]
```
